```python
#changing the name of old columns to new ones
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'state',
                       'region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed',
                       'nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                       'weaptype1_txt':'Weapon_type','motive':'Motive'}, inplace=True)
```

```python
#We only takes important columns which is requried in our dataframe.
df1=df[['Year','Month', 'Day','Country','state','city',
                       'Region','AttackType','Target','Killed','longitude', 'latitude',
                       'Wounded','Summary','Group','Target_type',
                       'Weapon_type','Motive']]
```

```python
#top five rows of new dataframe
df1.head()
```

```python
df1.info()
```

```python
#Fill the null values of the particular field (Wounded & Killed)
df1['Wounded'] = df1['Wounded'].fillna(0)
df1['Killed'] = df1['Killed'].fillna(0)
```

```python
#Make new column (Casualities) with merginig two columns
df1['Casualities'] = df1['Wounded'] + df1['Killed']
```

```python
df1.info()
```

```python
#Unique values in the Year column.
year = df1['Year'].unique()
#Value Count of the Year How Many Attacks Happened Each Year
years_count = df1['Year'].value_counts(dropna=False).sort_index()
plt.figure(figsize =(18,10))
sns.barplot(x=year,y=years_count,palette="tab10")
plt.xticks(rotation=50)
plt.xlabel('Attacking Year', fontsize =15)
plt.ylabel('Number of Attack each year', fontsize=15)
plt.title('Attacks In Year', fontsize = 20)
plt.show()
```

```python
#we are using no of attacks in year but region wise that'why we use crosstab here
pd.crosstab(df1.Year, df1.Region).plot(kind='area', stacked=False, figsize=(20,10))
plt.title('Terrorist Activities by Region in Each Year', fontsize = 20)
plt.ylabel('Number of Attacks', fontsize = 20)
plt.xlabel('Year', fontsize = 20)
plt.show()
```

```python
#In This we grouped the year and Killed Columns with year which gives us results
#of how many people die(Killed) in which year due to attack.
df2 = df1[['Year', 'Killed']].groupby(['Year']).sum()
fig, ax4 = plt.subplots(figsize=(20,10))
df2.plot(kind = 'bar', alpha=0.7, ax=ax4)
plt.xticks(rotation=50)
plt.ylabel('Number of killed People', fontsize=20)
plt.xlabel('Year', fontsize=20)
plt.title('People Die Due to Attacks', fontsize=20)
top_side = ax4.spines["top"]
top_side.set_visible(False)
right_side = ax4.spines["right"]
right_side.set_visible(False)
```

```python
#When we want city most affected then we use value count of city only.
df1['city'].value_counts().to_frame().sort_values('city', axis=0, ascending=False).head(10).plot(kind='bar', figsize=(20,10), color='yellow')
plt.xticks(rotation=50)
plt.xlabel('City', fontsize=20)
plt.ylabel('Number of Attacks', fontsize=20)
plt.title('Top 10 most effected city', fontsize=25)
plt.show()
```

```python
#when we want to show type of attacks then we just value counts the AttackType column
df1['AttackType'].value_counts().plot(kind='bar', figsize=(20,10), color='orange')
plt.xlabel('Attack Type', fontsize=15)
plt.ylabel('Number of Attack', fontsize=15)
plt.title('Name of Attack Type', fontsize=25)
plt.xticks(rotation=50)
plt.show()
```

```python
#In This we grouped the AttackType and Killed Columns with AttackType which gives us results
#of how many people die(Killed) due to which type of wepon.
df1[['AttackType','Killed']].groupby(["AttackType"],axis=0).sum().plot(kind='bar',figsize=(20,10),color=['darkslateblue'])
plt.xticks(rotation=50)
plt.title("Number of killed ",fontsize=20)
plt.ylabel('Number of people',fontsize=15)
plt.xlabel('Attack type',fontsize=15)
plt.show()
```

```python
#In This we grouped the AttackType and Wounded Columns with AttackType which gives us results
#of how many people Wounded due to which type of wepon.
df1[['AttackType','Wounded']].groupby(["AttackType"],axis=0).sum().plot(kind='bar',figsize=(20,10),color=['cyan'])
plt.xticks(rotation=50)
plt.title("Number of wounded  ",fontsize=20)
plt.ylabel('Number of people',fontsize=15)
plt.xlabel('Attack type',fontsize=15)
plt.show()
```

```python
plt.subplots(figsize=(20,10))
sns.countplot(df1["Target_type"],order=df1['Target_type'].value_counts().index,palette="gist_heat",edgecolor=sns.color_palette("mako"));
plt.xticks(rotation=90)
plt.xlabel("AttackType",fontsize=15)
plt.ylabel("count",fontsize=15)
plt.title("Attack per year",fontsize=20)
plt.show()
```

```python
df1['Group'].value_counts().to_frame().drop('Unknown').head(10).plot(kind='bar',color='green',figsize=(20,10))
plt.title("Top 10 terrorist group attack",fontsize=20)
plt.xlabel("terrorist group name",fontsize=15)
plt.ylabel("Attack number",fontsize=15)
plt.show()
```

```python
#In This we grouped the Group and Killed Columns with Group which gives us results
#of how many people die due by which terriorist group in attack
df1[['Group','Killed']].groupby(['Group'],axis=0).sum().drop('Unknown').sort_values('Killed',ascending=False).head(10).plot(kind='bar',color='yellow',figsize=(20,10))
plt.title("Top 10 terrorist group attack",fontsize=20)
plt.xlabel("terrorist group name",fontsize=15)
plt.ylabel("No of killed people",fontsize=15)
plt.show()
```
